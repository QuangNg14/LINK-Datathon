```python
billData = pd.read_excel("Purchase_Card_Transactions 2.xlsx")
billData = pd.DataFrame(billData)

billData.head()
```

```python
billData["TRANSACTION_DATE"][0]
```

```python
billData['AGENCY'].value_counts()
```

```python
billData['TRANSACTION_DATE'] = pd.to_datetime(billData['TRANSACTION_DATE'])
billData['TRANSACTION_DATE_2'] = pd.to_datetime(billData['TRANSACTION_DATE']).dt.date
billData["month"] = billData['TRANSACTION_DATE'].apply(lambda time: time.month)
billData["year"] = billData['TRANSACTION_DATE'].apply(lambda time: time.year)

billData.set_index(['TRANSACTION_DATE'], inplace = True)

billData = billData.sort_values(by = ['AGENCY', "TRANSACTION_DATE"])
```

```python
billData["Diff"] = billData["TRANSACTION_AMOUNT"].diff().fillna(0)
billData.head(100)
```

```python
#mean transaction amount of each agency
groupAgency_year = billData.groupby(["AGENCY"])["TRANSACTION_AMOUNT"].resample("Y").mean().fillna(0)
groupAgency_month = billData.groupby(["AGENCY"])["TRANSACTION_AMOUNT"].resample("M").mean().fillna(0)

df_year = groupAgency_year.to_frame()
df_month = groupAgency_month.to_frame()

df_year
df_month.to_csv("monthly spent.csv") 
df_year.to_csv("yearly spent.csv") 


# groupAgency.TRANSACTION_AMOUNT.resample("Y").mean()
```

```python
groupAgency_diff = billData.groupby(["AGENCY"])["Diff"].resample("M").mean().fillna(0)
groupAgency_diff = groupAgency_diff.to_frame()

groupAgency_diff.to_csv("monthly time frequency.csv")
```
